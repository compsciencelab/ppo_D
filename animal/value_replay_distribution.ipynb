```python
import matplotlib.pyplot as plt
import numpy as np
import random

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
trajectories = [i for i in range(50)]

values = [2 +(random.random()-1)-i*0.05 for i in trajectories]
ax.bar(trajectories,values)
plt.show()
```

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
pos_values = [abs(min(values)) + i for i in values]
pos_values = [i**10 for i in pos_values]
P = [i/sum(pos_values)  for i in pos_values]
ax.bar(trajectories,P)
plt.show()
```
